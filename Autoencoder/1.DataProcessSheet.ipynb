# Load Lab Data
All data from lab are pre-stored in class `LocalLab`, call it with `.data` 

```python
from utilities.dataloader import LocalLab, LocalChopchop
from sklearn.model_selection import train_test_split

seed = 546

# load lab dataset
lab = LocalLab('data/danrer11_guide_results.txt')

lab.data.to_csv("data/danrer11_lab_all.csv")

X_train, X_test = train_test_split(lab.data, test_size=0.2,random_state = seed)
X_train.to_csv("data/danrer11_lab_train.csv")
X_test.to_csv("data/danrer11_lab_test.csv")
```

# Load Chopchop Data

```python
chopchop = LocalChopchop("data/danrer11_chopchop.csv")

chopchop.data.to_csv("data/danrer11_chopchop_all.csv")

X_train, X_test = train_test_split(chopchop.data, test_size=0.2,random_state = seed)
X_train.to_csv("data/danrer11_chopchop_train.csv")
X_test.to_csv("data/danrer11_chopchop_test.csv")
```
